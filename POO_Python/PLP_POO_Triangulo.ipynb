```python
class Triangulo:

  def __init__(self, base, altura):
    self.base = base
    self.altura = altura

  def calcular_area(self):
    return self.base * self.altura

  def calcular_perimetro(self):
    return self.base*3

## Criando objetos e suas isnt√¢ncias

triangulo_1 = Triangulo(17, 9)
triangulo_1.calcular_area()
triangulo_1.calcular_perimetro()

print("Area = ", triangulo_1.calcular_area())
print("Perimetro = ", triangulo_1.calcular_perimetro())
```
