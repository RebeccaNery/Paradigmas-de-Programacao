```python
## Em Python, a sobrecarga de operadores permite que você defina comportamentos personalizados para operadores como +, -, *, etc., quando usados com objetos de uma classe definida por você. Isso é feito implementando métodos especiais dentro da sua classe. Esses métodos são conhecidos como métodos mágicos ou métodos dunder (double underscore).

class Produto:
  def __init__(self, nome, preco):
    self.nome = nome
    self.preco = preco

  def __str__(self):
    return(f"Produto: {self.nome} / Preço: R${self.preco}")

  def soma(self, outro):
    return f"{self.preco} + {outro.preco} = {self.preco + outro.preco}"

## Instâncias

produto1 = Produto("Shampoo", 10)
produto2 = Produto("Condicionador", 12)

## TESTES >>>

print(produto1)
print(produto2)

produto1.soma(produto2)
```
