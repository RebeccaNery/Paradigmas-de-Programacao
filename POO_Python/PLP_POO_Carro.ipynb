```python
class Carro:

## Método construtor da classe -- chamado automaticamente quando um objeto é instanciado e inicializa seus atributos
  def __init__(self, marca, modelo, ano, velocidade):
    self.marca = marca
    self.modelo = modelo
    self.ano = ano
    self.velocidade = velocidade

  def muda_velocidade(self, vel_nova):
    self.velocidade = vel_nova

  def exibir_velocidade(self):
    print("Velocidade:", self.velocidade)

## OBJETOS
carro_1 = Carro("Renault", "Sandero", 2015, 80)
carro_2 = Carro("Toyota", "Etios", 2017, 90)
carro_3 = Carro("Citroen", "C3", 2020, 120)

n = input("Qual carro quer exibir? ")

vel_nova = input("Deseja mudar a velocidade desse carro? SIM > digitar nova velocidade; NÃO > digitar 0 ")


## Exibir informações
print("Carro", n, ">>>")

if n == '1':
  print("Marca:", carro_1.marca)
  print("Modelo:", carro_1.modelo)
  print("Ano:", carro_1.ano)
  carro_1.exibir_velocidade()
  if vel_nova != '0':
    carro_1.muda_velocidade(vel_nova)
    carro_1.exibir_velocidade()
if n == '2':
  print("Marca:", carro_2.marca)
  print("Modelo:", carro_2.modelo)
  print("Ano:", carro_2.ano)
  carro_2.exibir_velocidade()
  if vel_nova != '0':
    carro_2.muda_velocidade(vel_nova)
    carro_2.exibir_velocidade()
if n == '3':
  print("Marca:", carro_3.marca)
  print("Modelo:", carro_3.modelo)
  print("Ano:", carro_3.ano)
  carro_3.exibir_velocidade()
  if vel_nova != '0':
    carro_3.muda_velocidade(vel_nova)
    carro_3.exibir_velocidade()
```
