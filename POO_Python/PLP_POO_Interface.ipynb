```python
## Interface imprimível com o método imprimir

from abc import ABC, abstractmethod

## Definindo a interface 'Imprimivel' >>>

class Imprimivel(ABC):

  def imprimir(self):
    pass


class Documento(Imprimivel):
  def __init__(self, titulo, tipo):
    self.titulo = titulo
    self.tipo = tipo

  def imprimir(self): ## implementação OBRIGATÓRIA do método 'imprimir', herdado de imprimivel e abc
    print(f"Documento: {self.titulo}.")
    print(f"Tipo: {self.tipo}.")

class Relatorio(Imprimivel):
  def __init__(self, responsavel, tipo):
    self.resp = responsavel
    self.tipo = tipo

  def imprimir(self): ## implementação OBRIGATÓRIA do método 'imprimir', herdado de imprimivel e abc
    print(f"Relatório ==> Responsável: {self.resp}.")
    print(f"Tipo: {self.tipo}.")

## instâncias

documento1 = Documento("Divórcio", "Jurídico")
documento2 = Documento("Certidão de Nascimento", "Opcional")
relatorio1 = Relatorio("Diretor", "Planilha")
relatorio2 = Relatorio("Coordenador", "Texto")

## TESTES >>>>

documento1.imprimir()
documento2.imprimir()
relatorio1.imprimir()
relatorio2.imprimir()

```
