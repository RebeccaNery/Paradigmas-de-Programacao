```python
class Retangulo:

  ## função que cria a Classe e já define seus atributos
  def __init__(self, comprimento, largura):
    self.comprimento = comprimento
    self.largura = largura

  def calcular_area(self):
    return self.comprimento * self.largura

  def calcular_perimetro(self):
    return 2* (self.comprimento + self.largura)

  ## Criando objetos e instâncias deles
ret = Retangulo(200, 300)
ret.calcular_area()
ret.calcular_perimetro()

print("Area: ",  ret.calcular_area())
print("Perimetro: ",  ret.calcular_perimetro())
```
