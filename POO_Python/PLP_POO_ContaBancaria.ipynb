```python
from PLP_POO_Exceptions import SaldoInsuficienteException

class ContaBancaria:

  def __init__(self, saldo, titular):
    self._saldo = saldo ##transformando o saldo em um atributo protegido
    self.titular = titular

  def depositar(self, valor):
    self._saldo = self._saldo + valor
    print("DepÃ³sito confirmado. Saldo atual:", self._saldo)

  def sacar(self, valor):
    if valor > self._saldo:
      raise SaldoInsuficienteException("Saldo Insuficiente.")
    self._saldo = self._saldo + valor
    print("Saque confirmado. Saldo atual:", self._saldo)


## Objetos
conta1 = ContaBancaria(1000, "Rebecca")
conta1.depositar(100)
conta1.sacar(2000)
```
