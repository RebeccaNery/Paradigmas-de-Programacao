```python
## implementar a classe Motor e associar à classe Carro, sendo que a classe carro precisa ter um objeto do tipo motor como um de seus atributos
## a relação aqui será de COMPOSIÇÃO, uma vez que o carro "possui" o motor; assim, se o a instância do objeto do tipo Carro for destruída, o objeto Motor associado a esse carro também é excluído da memória --> assim, se for desejado criar uma relação de composição, a instância do motor não pode ser criada separadamente à do  carro, mas dentro dela. Se for criada de forma separada, a relação é de associação.

#Diferença na memória:
#Composição: O objeto Motor é destruído quando o Carro é destruído.
#Agregação: O objeto Motor continua existindo mesmo depois de o Carro ser destruído, porque ele foi criado separadamente e é apenas referenciado pelo carro.
#Em resumo:
#Composição: O motor será apagado junto com o carro.
#Agregação: O motor será apenas desassociado do carro, mas continuará existindo na memória se houver outras referências a ele.

class Motor:
  def __init__(self, tamanho, combustivel):
    self.tamanho = tamanho
    self.combustivel = combustivel

  def __str__(self): # imprime na tela o tamanho e o combustível do motor
    return f"O motor {self.tamanho} utiliza {self.combustivel} como combustível."

class Carro:
  def __init__(self, cor, ano, tam, comb):
    self.cor = cor
    self.ano = ano
    self.motor = Motor(tam, comb) ## atributo que recebe um objeto do tipo Motor, sendo criado dentro da instância do carro

  def __str__(self):
    return f"O carro é {self.cor}, do ano de {self.ano} e seu motor é ==> {self.motor}."

## instâncias dos objetos
carro1 = Carro("vermelho", 2010, "1.6", "gasolina")
carro1.__str__()
```
