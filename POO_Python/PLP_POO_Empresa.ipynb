```python
## Modelar classe Empresa que agregue uma lista de objetos da classe Empregado. Atributos da classe Empregado: nome, cargo e salário

class Empregado:
  def __init__(self, nome, cargo, salario):
    self.nome = nome
    self.cargo = cargo
    self.salario = salario

  def __str__(self):
    return f"Empregado ==> Nome: {self.nome} / Cargo: {self.cargo} / Salário: {self.salario}."

class Empresa:
  def __init__(self, nome, produto):
    self.nome = nome
    self.produto = produto
    self.empregados = []

  def add_empregado(self, empregado):
    if empregado not in self.empregados:
      self.empregados.append(empregado) ##adiciona uma instância de Empregado à lista de empregados daquela empresa

  def __str__(self):
    empregados_str = ", ".join([empregado.nome for empregado in self.empregados])
    return f"A empresa {self.nome} vende {self.produto} e agrega os seguintes funcionários: {empregados_str}"

## Instâncias

empregado1 = Empregado("Gabriel", "Analista", 10500)
empregado2 = Empregado("Rebecca", "Professora", 12000)
empresaX = Empresa("Mushulandia", "Mushus de Plástico")

## TESTES >>>>>

empresaX.add_empregado(empregado1) ##agregando os empregados à empresa
empresaX.add_empregado(empregado2) ##agregando os empregados à empresa

print(empregado1)
print(empregado2)
print(empresaX)
```
