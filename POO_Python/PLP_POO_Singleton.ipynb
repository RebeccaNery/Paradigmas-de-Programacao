```python
## O padrão de design Singleton é um dos padrões de criação mais conhecidos na engenharia de software. Ele é usado para garantir que uma classe tenha apenas uma instância e fornecer um ponto global de acesso a essa instância. É útil quando você precisa exatamente um objeto para coordenar ações em todo o sistema.

class Configuracao:
  _instancia = None #atributo que armazena a instância única; inicialmente fica None para indicar que nenhuma instância foi criada ainda

  def __new__(cls, *args, **kwargs):
    if cls._instancia is None: # verifica se a instância já foi criada. Se não, então cria utilizando o método __new__ >>>
      cls._instancia = super(Configuracao, cls).__new__(cls, *args, **kwargs)

## Testando o Singleton

s1 = Configuracao()
s2 = Configuracao()

print(s1 is s2) ## True, pois ambos são exatamente a mesma instância
```
