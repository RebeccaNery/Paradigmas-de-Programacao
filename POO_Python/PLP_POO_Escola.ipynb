```python
## criar uma classe Escola e uma classe Professor, associando suas instâncias de forma que uma escola possa ter vários professores e um professor possa lecionar em várias escolas.
## a relação será de ASSOCIAÇÃO, pois as instâncias existirão de forma independente

class Escola:
  def __init__(self, nome, local):
    self.nome = nome
    self.local = local
    self.professores = []

  def add_prof(self, prof):
    if prof not in self.professores:
      self.professores.append(prof) ##adiciona uma instância de Professor à lista de professores daquela escola

  def __str__(self):
    professores_str = ",".join([prof.nome for prof in self.professores])
    return f"A Escola {self.nome} fica localizada em {self.local}. Lista de professores: {professores_str}"

class Professor:
  def __init__(self, nome, materia):
    self.nome = nome
    self.materia = materia
    self.escolas = []

  def add_escola(self, esc):
    if esc not in self.escolas:
      self.escolas.append(esc) ##adiciona uma instância de Escola à lista de escolas daquele professor

  def __str__(self):
    escolas_str = ", ".join([esc.nome for esc in self.escolas])
    return f"O professor {self.nome} leciona {self.materia} nas seguintes escolas: {escolas_str}"

## Instâncias dos Objetos
professor1 = Professor("Irazê", "Química")
professor2 = Professor("Mourão", "Matemática")
escola1 = Escola("Motiva", "Miramar")
escola2 = Escola("New Way", "Brisamar")
escola3 = Escola("Evolução", "Tambauzinho")

## Associando as instâncias

professor1.add_escola(escola1)
escola1.add_prof(professor1)
professor1.add_escola(escola3)

## TESTES
print(professor1)
print(professor2)
print(escola1)
print(escola2)
```
