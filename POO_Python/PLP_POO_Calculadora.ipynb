```python
## Na implementação da classe Calculadora, o método __init__ não foi necessário porque a classe não precisa inicializar ou armazenar nenhum atributo específico para as operações realizadas. Todos os métodos (adicionar, subtrair, etc.) recebem diretamente os parâmetros que são necessários para os cálculos, e não há necessidade de manter estado interno na classe.

## O método __init__ é o construtor da classe em Python. Ele é chamado automaticamente quando um objeto é instanciado e é usado para inicializar os atributos da instância. Você usa __init__ quando precisa que o objeto da classe armazene algum valor ou estado.

class Calculadora:
  def somar2(self, a, b):
    return f"Soma de {a} + {b} = {a+b}"

  def somar3(self, a, b, c):
    return f"Soma de {a} + {b} + {c} = {a+b+c}"

  def subtrair(self, a, b):
    return f"Subtração de {a} - {b} = {a-b}"

  def dividir(self, a, b):
    return f"Divisão de {a} por {b} = {a/b}"

  def multiplicar(self, a, b):
    return f"Multiplicação de {a} por {b} = {a*b}"

calculo1 = Calculadora()

calculo1.somar2(10,5)

#calculo1.somar3(1,2,3)
```
