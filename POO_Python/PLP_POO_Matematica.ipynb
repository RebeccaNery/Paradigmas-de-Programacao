```python
## Método estático

class Matematica:
  @staticmethod
  def fatorial(numero):
    fat_num = 1
    for i in range (1, numero + 1): ## precisa colocar numero + 1 porque a iteração no range pararia no número n (sem executá-lo)
      fat_num *= i
      if i != numero:
        print (f"{i} *")
      if i == numero:
        print(i)
    return f"Fatorial de {numero} ==> {fat_num}"

## Instâncias

calculo1 = Matematica()

## TESTES >>>

calculo1.fatorial(5)
```
