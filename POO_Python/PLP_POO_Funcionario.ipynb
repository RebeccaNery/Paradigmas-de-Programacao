```python
## Em Python, uma classe abstrata é uma classe que não pode ser instanciada diretamente e é usada para definir uma interface comum para um grupo de subclasses. Ela permite que você defina métodos que devem ser implementados pelas subclasses, promovendo uma estrutura coerente e forçando a implementação desses métodos.
## Para definir uma classe abstrata em Python, você usa o módulo abc (Abstract Base Classes).

from abc import ABC, abstractmethod

class Funcionario(ABC):
  @abstractmethod
  def calcularSalario(self): ## Método que deverá ser implementado pelas subclasses obrigatoriamente
    pass

class FuncionarioHorista(Funcionario):
  def __init__(self, nome, horas, valor_hora):
    self.nome = nome
    self.horas = horas
    self.valor_hora = valor_hora

  def calcularSalario(self):
    salario = self.horas * self.valor_hora
    return (f"{self.nome} trabalha {self.horas} horas e sua hora vale R${self.valor_hora} ; assim, seu salário é R${salario}")

  def __str__(self):
    return(f"Funcionário(a): {self.nome} / Horas: {self.horas} / Valor da hora: {self.valor_hora}")

class FuncionarioAssalariado(Funcionario):
  def __init__(self, nome, nivel):
    self.nome = nome
    self.nivel = nivel

  def calcularSalario(self):
    if self.nivel == 1:
      salario = 1400
    if self.nivel == 2:
      salario = 2500
    if self.nivel == 3:
      salario = 3200
    return (f"{self.nome} está no nível {self.nivel}, recebendo, portanto, R${salario}")

  def __str__(self):
    return(f"Funcionário(a): {self.nome} / Nível: {self.nivel}")

## Instâncias

funcionario1 = FuncionarioHorista("Tarcila", 10, 25)
funcionario2 = FuncionarioAssalariado("Ruan", 1)

## TESTES >>>

print(funcionario1)
print(funcionario2)

funcionario1.calcularSalario()
funcionario2.calcularSalario()
```
