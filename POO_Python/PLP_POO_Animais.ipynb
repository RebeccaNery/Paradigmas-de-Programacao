```python
## Classe base = Animais
class Animais:

  def __init__(self, nome, tamanho):
    self.nome = nome
    self.tamanho = tamanho

  def som(self):
    print("O", self.nome, self.tamanho,"faz barulho!")

## Classe Cachorro que herda de Animais
class Cachorro(Animais):

  def __init__(self, nome, tamanho, raca):
    super().__init__(nome, tamanho) #inicializa a classe base animal
    self.raca = raca

  def som(self): ##sobrescrita de método
    print("O cachorro faz auau!")
    print(self.nome, self.raca)

## Classe Gato que herda de Animais
class Gato(Animais):

  def __init__(self, nome, tamanho, raca):
    super().__init__(nome, tamanho)
    self.raca = raca

  def som(self): ##sobrescrita de método
    print("O gato faz miau!")
    print(self.nome, self.raca)

## Método para imprimir o som de cada animal utilizando polimorfismo (sobrescrita)
def imprimir_som(animal):
  print(animal.som())

## Objetos
animal1 = Animais("gato", "pequeno")
animal2 = Animais("cachorro", "médio")

animal1.som()
animal2.som()

cachorro1 = Cachorro("Bela", "Pug", "pequeno")
cachorro1.som()

gato1 = Gato("Mushu", "Laranja","gordo")
gato1.som()

imprimir_som(animal1)
imprimir_som(animal2)
imprimir_som(cachorro1)
imprimir_som(gato1)
```
